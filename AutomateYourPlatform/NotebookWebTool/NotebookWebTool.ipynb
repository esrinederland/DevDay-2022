#### Read input parameters

```python
featureService = { 
    "url": "https://services.arcgis.com/nSZVuSZjHpEZZbRo/ArcGIS/rest/services/OSM_POI/FeatureServer/0",
}

nrOfFeatures = 100
context = {}
```

#### Get random points from Feature Service

```python
from random import choices
from arcgis.features import FeatureLayer
from arcgis.geometry import Envelope, Polygon
from arcgis.geometry.filters import intersects

featureLayer = FeatureLayer(featureService["url"])

## Create a Geometry Filter using the extent
if "extent" in context:

    envelope = Envelope(context["extent"])
    extentFilter = intersects(envelope.polygon)
else:
    extentFilter = None

## Query all OIDs and select a random subset with size 'nrOfFeatures'
allOIDs = featureLayer.query(return_ids_only=True, out_sr=102100, geometry_filter=extentFilter)["objectIds"]
randomOIDs = choices(allOIDs, k=nrOfFeatures)

outputFeatureService = featureLayer.query(object_ids=",".join([f"{oid}" for oid in randomOIDs]), out_sr=102100)
```

#### Write the output parameter

```python
# Inserted snippet writes output parameter file for web tool execution
import os
# Only run this snippet during web tool execution when env variable "ENB_JOBID" will be present
if "ENB_JOBID" in os.environ:
    out_param_name = "outputFeatureService"
    out_param_file= os.path.join(os.environ["ENB_JOBID"], "value_" + out_param_name + ".dat")
    with open(out_param_file, "w") as writer:
        writer.write(str(outputFeatureService))
```
